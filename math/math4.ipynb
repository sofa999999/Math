Постройте график первой и второй производной функции
(sin(2x+1))^5
на отрезке [-5; 5]

```python
import matplotlib.pyplot as plt
import numpy as np
import pylab
import math

from sympy import *

x = symbols('x')
```

```python
diff1 = diff(x**5 + 4*sin(2*x) + cos(3*x+3))
print(diff1)
```

```python
from scipy.misc import derivative
def f(x):
    return x**5 + 4*sin(2*x) + cos(3*x+3)
derivative(f, 1.0)
```

```python
diff2 = diff(diff1)
diff2
```

```python
def f2(x):
    return 5*x**4 - 3*sin(3*x + 3) + 8*cos(2*x)
derivative(f2, 1.0)
```

Постройте график первой и второй производной функции (sin(2x+1))^5 на отрезке [-5; 5]

```python
d1 = diff((sin(2*x+1))**5)
print(d1)
```

```python
d2= diff(d1)
print(d2)
```

```python

x = np.linspace(-5, 5)                   
y1 = (np.sin(2*x+1))**5                            
y2 = 10*np.sin(2*x + 1)**4*np.cos(2*x + 1)
y3 = -20*np.sin(2*x + 1)**5 + 80*np.sin(2*x + 1)**3*np.cos(2*x + 1)**2
fig, ax = plt.subplots()                       
ax.plot(x, y1, color="blue", label="y(x)")      
ax.plot(x, y2, color="red", label="y'(x)")      
ax.plot(x, y3, color="green", label="y''(x)")   
ax.set_xlabel("x")                              
ax.set_ylabel("y")                              
ax.legend()                                     

plt.show()                                     


```
