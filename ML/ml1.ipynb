```python
selectedColumns = data2[['age', 'workclass', 'education','marital-status', 'occupation', 'race', 'gender', 'hours-per-week', 'native-country', 'income']]
```

```python
X = pd.get_dummies( selectedColumns, columns = [ 'workclass', 'education', 'marital-status', 'occupation', 'race', 'gender', 'native-country' ] )
X.head()
```

```python
del X['income']
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit( data2['income'] )
```

```python
le.classes_
```

```python
le.transform( [ '<=50K', '>50K' ] )
```

```python
y = pd.Series( data = le.transform( data2['income'] ) )
y.head()
```

```python
model = LogisticRegression()
```

```python
model.fit( X, y )
predictions = model.predict_proba( X )
```

```python
predictions[:5]
```

```python
model.score(X, y)
```
